### Simplifiy Worms taxonomy data and only use "accepted" species for now

```python
cols = [
    "scientificName",
    "kingdom",
    "phylum",
    "class",
    "order",
    "family",
    "genus",
    # "subgenus",
    "specificEpithet",
    # "infraspecificEpithet",
    "taxonRank",
    # 'taxonomicStatus'
    # "nomenclaturalStatus",
]

worms_taxon_simplified = worms_taxon.query('taxonomicStatus == "accepted"')[cols]
print(worms_taxon_simplified.shape)
worms_taxon_simplified.head()
```

## Worms identifiers (mapping to NCBI taxonomy)

```python
worms_identifiers = pd.read_csv(
    "/Users/olgabot/botryllus-data/WoRMS_download_2024-04-01/identifier.txt",
    sep="\t",
    index_col=0,
)
print(worms_identifiers.shape)
worms_identifiers.head()
```

### Cut down to just a few columns for simplified joining

```python
worms_identifiers.shape
```

```python
worms_identifiers_simplified = worms_identifiers.query('datasetID == "ncbi"')[
    ["ncbi_taxid", "title"]
]  # .reset_index()
print(worms_identifiers_simplified.shape)
worms_identifiers_simplified.head()
```

## Worms Marine/Terrestrial info

```python
worms_species_profile = pd.read_csv(
    "/Users/olgabot/botryllus-data/WoRMS_download_2024-04-01/speciesprofile.txt",
    sep="\t",
    index_col=0,
)
print(worms_species_profile.shape)
worms_species_profile.head()
```

## Join WORMS data together

```python
worms_species_profile_ids = (
    worms_species_profile.join(worms_identifiers_simplified, how="inner")
    .join(worms_taxon_simplified, how="inner")
    .reset_index()
)
print(worms_species_profile_ids.shape)
worms_species_profile_ids.head()
```

## Join with Eukaryotes, add column if any kind of aquatic (marine, freshwater, brackish)

```python
eukaryota_counted_worms = eukaryota_counted.merge(
    worms_species_profile_ids,
    how="left",
    left_on="Organism (ID)",
    right_on="ncbi_taxid",
)

water_cols = ["isMarine", "isFreshwater", "isBrackish"]

eukaryota_counted_worms["any_marine_freshwater_brackish"] = (
    eukaryota_counted_worms[water_cols].any(axis=1)
)
print(eukaryota_counted_worms["any_marine_freshwater_brackish"].sum())
print(eukaryota_counted_worms.shape)
eukaryota_counted_worms.head()
```

### Save to file!

```python
eukaryota_counted_worms.to_parquet(
    "/Users/olgabot/botryllus-data/2024-04-15__uniprot_eukaryota_counted_per_kingdom__worms_ismarine.parquet"
)
```

### Subset to only Reviewed

```python
eukaryota_counted_worms_reviewed = eukaryota_counted_worms.query(
    "Reviewed == 'reviewed'"
)
print(eukaryota_counted_worms_reviewed.shape)
eukaryota_counted_worms_reviewed.head()
```

### Save to file!

```python
eukaryota_counted_worms_reviewed.to_parquet(
    "/Users/olgabot/botryllus-data/2024-04-15__uniprot_eukaryota_counted_per_kingdom__worms_ismarine__reviewed.parquet"
)
```
