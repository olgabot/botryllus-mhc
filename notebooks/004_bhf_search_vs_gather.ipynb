### Create HP match dictionary

```python
hydrophobic_residues = [k for k, v in HP_MAPPING.items() if v == 'h']
polar_residues = [k for k, v in HP_MAPPING.items() if v == 'p']
polar_residues
```

```python
hydrophobic_hydrophobic = dict.fromkeys(itertools.product(hydrophobic_residues, hydrophobic_residues), 1)
polar_polar = dict.fromkeys(itertools.product(polar_residues, polar_residues), 1)

non_match = 0

hydrophobic_polar = dict.fromkeys(itertools.product(hydrophobic_residues, polar_residues), non_match)
polar_hydrophobic = dict.fromkeys(itertools.product(polar_residues, hydrophobic_residues), non_match)

hp_match_dictionary = dict()
hp_match_dictionary.update(hydrophobic_hydrophobic)
hp_match_dictionary.update(polar_polar)
hp_match_dictionary.update(hydrophobic_polar)
hp_match_dictionary.update(polar_hydrophobic)
len(hp_match_dictionary)
```

```python
for a in pairwise2.align.localdx(tnpo2['sequence'], bhf['sequence'], hp_match_dictionary, one_alignment_only=True):
    print(format_alignment(*a))
```
