```python
%%time
import pandas as pd

eukaryota = pd.read_csv(
    "/Users/olgabot/botryllus-data/2024-03-30__uniprot_eukaryota.tsv.gz",
    sep="\t",
    nrows=1000,
)
eukaryota["opisthokonta"] = eukaryota["Taxonomic lineage"].str.contains("Opisthokonta")
eukaryota["metazoa"] = eukaryota["Taxonomic lineage"].str.contains("Metazoa")
eukaryota["fungi"] = eukaryota["Taxonomic lineage"].str.contains("Fungi")
eukaryota["viridiplantae"] = eukaryota["Taxonomic lineage"].str.contains(
    "Viridiplantae"
)
eukaryota.head()
```

```python
eukaryota.groupby(["Organism", "Organism (ID)", "Reviewed"]).size()
```

```python
%%time

eukaryota_counted = eukaryota.groupby(
    [
        "opisthokonta",
        "metazoa",
        "fungi",
        "viridiplantae",
        "Organism",
        "Organism (ID)",
        "Taxonomic lineage",
        "Reviewed",
    ]
).size()
eukaryota_counted
```
